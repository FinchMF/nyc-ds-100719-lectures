**Aim: SWBAT create a figure with 4 subplot of varying graph types.**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy.random import seed, randint
seed(100)

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=True, dpi=120)

# Define the colors and markers to use
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Plot each axes
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('Four Subplots in One Figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
# plt.show()
```

```python
# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=False, dpi=200)

# Define the colors and markers to use
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Plot each axes
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('Four Subplots in One Figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
```

### Go through and play  with the code above to try answer the questions below:

- What do you think `sharex` and `sharey` do?    

formatting tool that allows a consistency x and y axes between all the subplots


- What does the `dpi` argument control?

dpi means Dots Per Inch. It controls the size of the entire plot

- What does `numpy.ravel()` do, and why do they call it here?

numpy.ravel() enumerates the information via for loop into the charts

- What does `yaxis.set_ticks_position()` do?

positions the ticks on the yaxis amoung the subplots


- How do they use the `colors` and `markers` dictionaries?

the colors and markers are sequentially looped, right to left, through the dictionary so each key:value has their own designated color and marker. Using the same key between two dictionaries and setting up a for loop to be enumerated, can assign information from seperate dictionaries onto a shared graph.

```python
from numpy.random import seed, randint
seed(100)



x = sorted(randint(0,10,10))
x2 = sorted(randint(0,20,10))
y = sorted(randint(0,10,10))
y2 = sorted(randint(0,20,10))

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=False, sharey=False, dpi=120)

# Define the colors and markers to use
colors = {0:'b', 1:'b', 2:'b', 3:'b'}
markers = {0:'o', 1:'o', 2:'o', 3:'o'}


plt.scatter(x,y, label = 'skiscat', color = 'r', s=15, marker = 'o')
 
    
axes[0][0].plot(x2,y2, color = 'g')
axes[0][1].scatter(x,y, color = 'g')
axes[1][0].scatter(x,y, color = 'g')
axes[1][1].plot(x2,y2, color = 'g')

# Plot each axes
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('none')
    axes[0][0].plot(x,y, color = 'aquamarine')
    axes[0][1].scatter(x2,y2, color = 'aquamarine')
    axes[1][0].scatter(x2,y2, color = 'aquamarine')
    axes[1][1].plot(x,y, color = 'aquamarine')
fig

plt.suptitle('Four New Subplots in One Figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
# plt.show()
```
